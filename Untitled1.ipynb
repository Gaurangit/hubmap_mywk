```python
def nodes_d(nodes_data):
    dict_nodes={}
    for nn in nodes_data:
        
        dict_nodes.update({nn['id']: Node(nn['type'], name=nn['name'], onto_id=nn['metadata']['ontologyId'], label=nn['metadata']['label'])})
    return dict_nodes

def create_relation(edges_data):
    for edge in edges_data:
        graph.create(Relationship(dict_nodes[edge['target']], "is_part_of", dict_nodes[edge['source']]))
```

```python
##Creating nodes
dict_nodes=nodes_d(data['nodes'])
for i in dict_nodes.keys():
    graph.create(dict_nodes[i])
    
    
    
create_relation(data['edges'])
```
