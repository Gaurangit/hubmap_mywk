```python
import pandas as pd
import numpy as np
import os
```

```python
organ_name='kidney'


dff=pd.DataFrame(columns=['organ', 'tissue_type', 'cell_type', 'cell_type_ontology_id', 'disease',
       'sex', 'ethnicity', 'development_stage', 'count'])
#arr=dff.values
for file in os.listdir('non_gene_all_47/'):

    try:
        df=pd.read_excel("non_gene_all_47/"+file)
        if organ_name in set(df['organ']):
            print(file)
            #print(list(df['organ']).count('kidney'))
            #arr = np.vstack((arr, df.values))
            dff=pd.concat([dff,df])
    except:
        print("bad_file -->", file)
```

```python
for org, data in dff.groupby("organ"):
    #print(org)
    if org==organ_name:
        df_organ=data
```
